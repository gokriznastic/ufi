```python
from itertools import permutations
import random

test_filelist = {}
train_filelist = {}
pairs = []

for i in range(1, 606):
    train_filelist['s'+str(i)] = ['data/train/s'+str(i)+'/'+f for f in os.listdir('data/train/s'+str(i))]
for i in range(1, 606):
    test_filelist['s'+str(i)] = ['data/test/s'+str(i)+'/'+f for f in os.listdir('data/test/s'+str(i))]

for key in test_filelist.keys():
    for f in train_filelist[key]:
        name = f.split('/')[2]
        file1 = f.split('/')[3][:-4]
        file2 = test_filelist[key][0].split('/')[3][:-4]
        pairs.append(name + ' ' + file1 + ' ' + file2)
```
